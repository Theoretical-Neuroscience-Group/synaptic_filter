```python
def benchmark_simulation(p, online=True, verbose = False, precomputed_generator=None):

    # prepare
    out = []
    t_stop = int(p['tau_ou']//p['dt']*p['epoch_num'])
    t_readout = int(p['tau_ou']*p['epoch_wait']//p['dt'])
    if online:
        t_num = 2
    else:
        p['t_num'] = t_stop + 1
        t_num = p['t_num']

    # run
    v = init(p,t_num = t_num)
    assert 'x_wiggle' in v

    # update w, Sx, g, y
    if precomputed_generator is not None:
        keys = ['Sx', 'y', 'w', 'g']
        for key in keys:
            v[key] = precomputed_generator[key]
        errors = {}
        print('loaded precomputed values for:',keys)

    for t in range(1):
        k = 0 if online else t

        if precomputed_generator is None:
            tic = 1000*time.perf_counter()
            errors = update_generator(v,p,k) # ground truth weights, spikes
            toc = 1000*time.perf_counter()
            print(f"update_generator: {toc - tic:0.6f} ms")
        
        tic = 1000*time.perf_counter()
        update_protocol(v,p,k)  # compute kernels
        toc = 1000*time.perf_counter()
        print(f"update_protocol: {toc - tic:0.4f} ms")
        
        tic = 1000*time.perf_counter()
        update_filter(v,p,k)
        toc = 1000*time.perf_counter()
        print(f"update_filter: {toc - tic:0.4f} ms")

        if online is True:
            shift_vars_back(v,k)

        if t >= t_readout:
            res = get_performance(v,p,k=k)
            res.update(errors) # include errors in output
            out.append(res)

        if ((verbose is True) and (t % int(p['tau_ou']//p['dt']) == 0)):
            # progresss
            print(np.ceil(100*t/t_stop), end='\r')

    return pd.DataFrame(out), v
```

```python
p['rule'] = 'corr'
out,v2 = benchmark_simulation(p,verbose=True,online=False)
```

```python
p['rule'] = 'exp'
out,v2 = benchmark_simulation(p,verbose=True,online=False)
```
